## Juego 'Adivina el número'

Se pretende realizar mediante el lenguaje Python un programa que simule un juego de 
adivinanza.

Al comenzar el juego y durante el desarrollo del mismo las opciones que se muestran 
hasta elegir la opción salir son:
  1. Partida modo solitario
  2. Partida 2 Jugadores
  3. Estadística
  4. Salir

Modo solitario supone que el número a adivinar es generado aleatoriamente por el 
ordenador.
Modo 2 Jugadores: primero se escribe un número a adivinar y luego un segundo jugador 
intenta adivinarlo.

El número a adivinar debe ser entre 1 y 1000.

Tanto la opción 1 como la opción 2 tendrán el siguiente submenú para elegir la 
dificultad:
  1. Fácil (20 intentos)
  2. Medio (12 intentos)
  3. Difícil (5 intentos)

En ambos menús debe chequearse que la opción elegida es válida y avisar en caso 
contrario.
Si se adivina el número se gana la partida y se vuelve al menú principal, sino se indica 
si el número buscado es mayor o menor.
Si se supera el número de intentos se pierde y se vuelve al menú principal.
Se gane o se pierda se pide el nombre del jugador y se guarda esta información junto 
con el resultado de la partida en un fichero Excel.
La opción estadística nos mostrara los datos almacenados en Excel.

```python
import random
import getpass
import time
from colorama import init, Fore, Back, Style
import openpyxl
import pandas as pd

# 1. Elegir modo de juego(a través de input opcion):

opcion = ''
numero = 1

def modo_juego():
    global opcion
    global numero
    opcion = input(Fore.BLUE+'Elige el modo de juego: \n 1. Partida modo solitario \n 2. Partida 2 Jugadores \n 3. Estadística \n 4. Salir \n'+Fore.RESET)
    if opcion == '1':
        time.sleep(0.5)
        print('\nHas elegido la opcion ' + opcion + ', la partida en modo Solitario.')
        numero = random.randint(1, 1000)
    elif opcion == '2':
        time.sleep(0.5)
        print('\nHas elegido la opcion ' + opcion + ', la partida en modo Multijugador. \n')
        time.sleep(0.5)
        numero = int(getpass.getpass(Fore.MAGENTA +'Jugador 1, Introduce el número a adivinar por el Jugador 2. Debe estar entre 1 y 1000: '+Fore.RESET))
        while numero < 1 or numero > 1000:
            time.sleep(1)
            print(Fore.RED+'Jugador 1, el número debe estar entre 1 y 1000.'+ Fore.RESET)
            time.sleep(0.5)
            numero = int(getpass.getpass(Fore.MAGENTA+"Jugador 1, Introduce el número a adivinar por el Jugador 2. Debe estar entre 1 y 1000: "+Fore.RESET))
            time.sleep(1)
        print('\n El jugador 1 ha introducido el número a adivinar.')
    elif opcion == '3':
        time.sleep(0.5)
        print('\nHas elegido la opcion ' + opcion + '.')
        time.sleep(0.5)
        estadisticas()
    elif opcion == '4':
        print('\nHas elegido salir del juego.')
        time.sleep(1)
        print('\n¡Hasta la próxima!')
    else:
        time.sleep(0.5)
        print(Back.RED+'Opción inválida. Introduce una de las 4 opciones. '+ Back.RESET)
        time.sleep(1)
        modo_juego()
    return 
       
    
# 2. Elegir dificultad en las dos primeras opciones de juego(número de intentos):

intentos = ''

def elegir_dificultad():                
    global intentos
    dificultad = input(Fore.BLUE+'Ahora elige una dificultad: \n 1. Fácil (20 intentos) \n 2. Medio (12 intentos) \n 3. Difícil (5 intentos) \n')
    if dificultad == '1' or dificultad == 'Fácil':
        intentos = 20
        time.sleep(0.5)
        print('\nHas escogido tener ' + str(intentos) + ' intentos.')
    elif dificultad == '2' or dificultad == 'Medio':
        intentos = 12
        time.sleep(0.5)
        print('\nHas escogido tener ' + str(intentos) + ' intentos.')
    elif dificultad == '3' or dificultad == 'Difícil':
        intentos = 5
        time.sleep(0.5)
        print('\nHas escogido tener ' + str(intentos) + ' intentos.')
    else:
        time.sleep(0.5) 
        print(Back.RED+'Dificultad inválida. Escoge una de las 3 dificultades.'+ Back.RESET)
        time.sleep(0.5)
        elegir_dificultad()
    return


# Funciones auxiliares internas para la principal de adivinar_numero:

def fallo(gap):
    time.sleep(0.5)
    print(Fore.RED +'\n ¡Fallaste!' + Fore.RESET)
    time.sleep(1)
    print('\nTe doy una pista: el número a adivinar es ' + gap + '.')
    time.sleep(1)
    print('\nSigue probando. Te quedan ' +str(intentos)+ ' intentos.\n')
    time.sleep(1)
    

def fin_intentos():
    time.sleep(1)
    print('\nOooohh, Se te han acabado los intentos. El número a adivinar era el ' + str(numero) + '.')
    time.sleep(1)    

    
# Creamos las variables en las que vamos a guardar los datos que van al excel
nombres = ''
resultados = '' 

def save_data():
    excel_document = openpyxl.load_workbook('C:\\Users\\tiqda\\tarea.xlsx')
    Hoja = excel_document['Resultados']
    Hoja.append([nombres, resultados])                 #introducimos los datos en una lista para que los convierta en columnas en el excel
    save_data = excel_document.save('C:\\Users\\tiqda\\tarea.xlsx')

def estadisticas():
    print('\n Creando estadísticas...\n ')
    time.sleep(1.5)
    df = pd.read_excel('C:\\Users\\tiqda\\tarea.xlsx', header = None)
    df.columns = ['Nombre', 'Resultado']               #cambiamos el nombre de las columnas del dataframe (no tenía)
    print(df)
        
def fin_partida():
    global nombres
    nombres = input(Fore.CYAN+'\nEscribe tu nombre para guardar la partida junto con la de otros jugadores: '+Fore.RESET)
    time.sleep(0.5)
    print('Muchas gracias, ' + nombres + '.')
    save_data()                                        #Guardamos los datos de la partida
    time.sleep(1)
    print('\n¿Jugamos de nuevo? \n')
    time.sleep(2)
    juego()   

    
# 3. Adivinar_numero en las dos primeras opciones del juego(función principal):

def adivinar_numero():
    global intentos
    global resultados
    while intentos > 0:
        tunumero = int(input(Fore.BLUE+'Jugador, Adivina el número entre 1 y 1000: '+Fore.RESET))
        if tunumero > numero:
            intentos -= 1
            if intentos > 0:
                gap = 'menor'
                fallo(gap)
            else:
                resultados = 'Pierde'
                fin_intentos()
                fin_partida
                juego()
        elif tunumero < numero: 
            intentos -= 1
            if intentos > 0:
                gap = 'mayor'
                fallo(gap)
            else:
                resultados = 'Pierde'
                fin_intentos()
                fin_partida()
                juego()
        elif tunumero == numero:
            time.sleep(1)
            print(Back.GREEN+'\n¡Enhorabuena, acertaste!'+Back.RESET)
            resultados = 'Gana'
            fin_partida()                            
            break
    return    


# 4. Definimos la función completa del juego:

def juego():
    print()
    modo_juego()
    while opcion == '1' or opcion == '2': 
        time.sleep(1)
        print('')
        elegir_dificultad()
        time.sleep(1)
        print('')
        adivinar_numero()
        if opcion == '4':
            break

# FINAL. EJECUTAMOS EL JUEGO:

print('\n¡Bienvenido al juego Adivina el número!\n')  #lo dejo aparte para que solo salga la primera vez que se ejecuta el juego
time.sleep(1)
juego()
```
